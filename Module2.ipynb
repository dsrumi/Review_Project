```python
import pandas as pd
import numpy as np
```

```python
column_names = ["sepal length", "sepal width", "petal length","petal width","class"]  
df = pd.read_csv("iris.data", names=column_names, header=None)
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Histogram for 'sepal length'
plt.figure(figsize=(6, 4))
sns.histplot(df['sepal length'], bins=20, kde=True)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart showing mean 'petal width' for each species
plt.figure(figsize=(6, 4))
sns.barplot(x='class', y='petal width', data=df, estimator=lambda x: x.mean())
plt.title('Mean Petal Width for Each Species')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width')
plt.show()
```

```python


# Scatter plot comparing 'sepal length' and 'sepal width' for each species
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal length', y='sepal width', hue='class', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```
